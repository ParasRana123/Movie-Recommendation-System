```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
import bs4 as bs
import urllib.request
```

### Extracting features of the 2023 Bollywood movies

```python
link = "https://en.wikipedia.org/wiki/List_of_Hindi_films_of_2023"
```

```python
source = urllib.request.urlopen(link).read()
soup = bs.BeautifulSoup(source , 'lxml')
```

```python
tables = soup.find_all('table' , class_='wikitable')
```

```python
len(tables)
```

```python
df1 = pd.read_html(str(tables[0]))[0]
df2 = pd.read_html(str(tables[1]))[0]
df3 = pd.read_html(str(tables[2]))[0]
df4 = pd.read_html(str(tables[3]))[0] # avoided "ValueError: invalid literal for int() with base 10: '1"'
df5 = pd.read_html(str(tables[4]))[0]
df6 = pd.read_html(str(tables))
```

```python
df = df1.append(df2.append(df3.append(df4 , ignore_index=True) , ignore_index=True) , ignore_index=True)
```
