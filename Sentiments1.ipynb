### Sentiment Analysis on a reviews.txt file using Bidirectional LSTM RNN

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('reviews.txt' , sep='\t' , names=['Reviews' , 'Comments'])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
# getting the independent feature
X = df.drop('Reviews' , axis=1)
```

```python
# getting the dependent features
y = df['Reviews']
```

```python
# Checking whether the dataset is balanced or not
y.value_counts()
```

```python
import tensorflow as tf
tf.__version__
```

```python
# Importing the Important Libraries
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Bidirectional
```

```python
# Fixing the vocabulary size
voc_size = 5000
```

```python
messages = X.copy()
```

```python
messages
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
# Now Implementing the Data Preprocessing using the Porter stemmer technique
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
for i in range(0, len(messages)):
    print(i)
    review = re.sub('[^a-zA-Z]', ' ', messages['Comments'][i])
    review = review.lower()
    review = review.split()
    
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
# One hot encoding of the Corpus that we just got
onehot_repr = [one_hot(words , voc_size) for words in corpus]
print(onehot_repr)
```

### Embedding Representation

```python
sent_length = 30
embedded_docs = pad_sequences(onehot_repr , padding='pre' , maxlen=sent_length)
embedded_docs
```

```python
# Creating the model...
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
from tensorflow.keras.optimizers import Adam

# Define the parameters
embedding_vector_features = 60
voc_size = 5000  # Example vocabulary size, change as needed
sent_length = 30  # Example sentence length, change as needed

# Create the model
model = Sequential()
model.add(Embedding(input_dim=voc_size, output_dim=embedding_vector_features, input_length=sent_length))
model.add(Bidirectional(LSTM(100)))
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(loss='binary_crossentropy', optimizer=Adam(), metrics=['accuracy'])

# Print the model summary
model.build(input_shape=(None, sent_length))  # Manually build the model with the input shape
print(model.summary())
```

```python
# Converting the embedded docs to a numpy array
X_final = np.array(embedded_docs)
y_final = np.array(y)
```

### Train-Test Split

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X_final , y_final , test_size=0.33 , random_state=42)
```

```python
# Training the Model....
model.fit(X_train , y_train , validation_data=(X_test , y_test) , epochs=10 , batch_size=32)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
import numpy as np
y_pred = np.where(y_pred>=0.5 , 1 , 0)
```

```python
from sklearn.metrics import confusion_matrix , accuracy_score
confusion_matrix(y_test , y_pred)
```

```python
print(accuracy_score(y_test , y_pred))
```
